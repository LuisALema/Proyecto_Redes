## 2. Redes LAN

## Distribucion de dispositivos
### Sede Rio de Janeiro
En esta sede los dispositivos dentro de una figura color naranja y verde serán monitoreados desde la sede de Seguridad y control.
En el switch del datacenter se cuenta con 2 vlans, uno para la sede de seguridad y otro para los dispositivos propios del datacenter, como el servidor web.
En la parte de Administracion y Coordinacion de igual forma se cuenta con dispotivos propios de las oficinas y tambien dispositivos para monitorear las operaciones internas.
### Sede Recife
En esta sede se harán los controles y monitoreos de sensores, camara, etc, además se contará con el servidor IOT que será el que controle los actuadores en base a las mediciones de los sensores.
El area de Seguridad será la encargada de monitorear las camarás de las sedes en las que se dispongan de las mismas, en cambio el area de control se encargará de controlar el ambiende de las areas en donde se cuente con dispositivos de actuacion.
### Sede Sao Paulo
Se tienen dispositivos de prensa, como camaras de grabacion y el dispositivos de visualizacion de imagen (laptop), ademas se implmentó un AP con el fin de tener la prensa acreditada conectada inalambricamente.
### Sede Salvador
Aqui se tiene los escenarios principales, se tiene dispositivos de vigilancia y de presencia controlados desde la sede Recife, con el fin de controlar el no acceso a areas sensibles, o visualizar alguna aglomeracion de ser el caso. 
Se tiene una bodega con los articulos de logistica, los cuales son monitoreados para mantener un ambiente correcto, ademas de evitar accesos a esta area.
Tamgien se cuenta con una oficina de logistica y produccion las cuales son las que recibe los comunicados desde las otras sedes.
Se cuenta con un AP para la conexion inalambrica de personal de campo y tecnicos de apoyo.
